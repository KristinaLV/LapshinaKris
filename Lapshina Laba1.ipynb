**Простейшая модель Мальтуса роста населения Земли:**
$$1)\ x' = kx$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(x,t):
    k=0.42
    return k * x
```

```python
t = np.linspace(0,10,100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
plt.plot(t, solve, label = r"$x' = kx$")
plt.grid(True)
plt.legend()
plt.show()
```

**Мягкая модель роста с зависящими от населения коэффициентом размножения**

$2)\ x' = k(x)x, \ k(x) = a - bx$

```python
def k(x):
    a=1
    b=4
    return a - b*x
def dydt(x,t):
    return k(x)*x
```

```python
t = np.linspace(0,10,100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
plt.plot(t, solve, label = r"$x' = k(x)x$")
plt.grid(True)
plt.legend()
plt.show()
```

**Модель роста популяции рыб**

$$3) \ x' = x - x^2 - c$$

```python
def dydt(x,t):
    c = -0.4
    return x - x*x - c
```

**Мягкая модель роста популяции рыб:**

$$4) \ x' = x - g(x)x - c,$$
