```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.arange(1,11)
y1=[-7, -3, 1, -2, 4, 6, -5, 11, 13, 17]

%matplotlib inline

plt.plot(x1, y1, ".", color="red", label = r"kris")
plt.grid(True)
plt.legend()
plt.show()
```

```python
X = np.vstack([x1, np.ones(len(x1))])
X = X.T
```

```python
lsq = np.linalg.lstsq(X, y1, rcond = None)
a,b=lsq[0]

%matplotlib inline

plt.plot(x1, y1, ".")
plt.plot(x1, a*x1 + b)
```

```python
from scipy.optimize import curve_fit

def f(x1, a, b):
    return a * x1**2 + b * x1
(a, b), _ = curve_fit(f, x1, y1)

%matplotlib inline

plt.plot(x1, y1, ".")
plt.plot(x1, f(x1, a, b))
```

```python
from sklearn.preprocessing import PolynomialFeatures

from sklearn.pipeline import make_pipeline

from sklearn.linear_model import Ridge
```

```python
x_m = x1[:, np.newaxis]
model = make_pipeline(PolynomialFeatures(1), Ridge())
model.fit(x_m, y1)
```

```python
x_long = np.linspace(1, 11, 100)[:, np.newaxis]
y_p = model.predict(x_long)

plt.plot(x_long, y_p)
plt.plot(x1, y1, ".", color="red", label = r"kris")
```
