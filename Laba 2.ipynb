```python

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
    a=400000000
    b=350
    return a - b*x

def dydt(x,t):
    return k(x)*x

solve = odeint(dydt, y0=4000, t=years)
solve = solve.flatten()

plt.plot(years, solve, label = r"Жёсткая модель", color = "green")
plt.plot(years, people, color="red", label = r"Население Екатеринбурга")
plt.grid(True)
plt.legend()
plt.show()
```
